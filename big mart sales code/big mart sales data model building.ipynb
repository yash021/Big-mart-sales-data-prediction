```python
import numpy as np
import pandas as pd
new_train=pd.read_csv("F:\\ML\\jupyter notebook\\train_modified.csv")
new_test=pd.read_csv("F:\\ML\\jupyter notebook\\test_modified.csv")
```

```python
x_train=new_train.drop(['Item_Outlet_Sales'],axis=1)
y_train=new_train.iloc[:,2].values
```

```python
x_train
```

```python
x_train=x_train.drop(['Item_Identifier','Outlet_Identifier'],axis=1)
new_test=new_test.drop(['Item_Identifier','Outlet_Identifier'],axis=1)
```

```python
x_train
```

```python
# Fitting simple regression to the training set
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Predicting the test results
y_pred=regressor.predict(new_test)
y_pred
```

```python
y_pred=pd.DataFrame(y_pred)
y_pred
```

```python
y_pred.to_csv("y_test.csv",index=False)
```
